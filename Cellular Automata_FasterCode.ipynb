```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def CAInitial(time):
    "time = m are the rows, sites = n = 2*m+1" 
    
    m = time 
    n = 2*m+3
    "+3 because this inlcudes 2 sites to implement the fixed boundary conditions"
    CA = np.zeros((m,n),dtype=int)
    CA[0, m] = 1 
    
    return CA



def CARule(N):
    "takes number N=0,...255 as input and translates it into a local CA rule"
    "we create a dictionary CArule that maps local environment onto new state according to rule encoded in N"
    
    binString= bin(N)[2:].zfill(8)
    "this returns N as an 8bit number"
    
    R = {}
    R[1,1,1] = binString[0]
    R[1,1,0] = binString[1]
    R[1,0,1] = binString[2]
    R[1,0,0] = binString[3]
    R[0,1,1] = binString[4]
    R[0,1,0] = binString[5]
    R[0,0,1] = binString[6]
    R[0,0,0] = binString[7]
    
    return R



def CAEvolution(time,rule):
    "this created the CA for time t according to rule"
    
    N = rule
    m = time 
    n = 2*m + 3
            
    CA = CAInitial(m)

    for t in range(1, m):
        
        for site in range(1,n-1):
            
            neighbor=tuple(CA[t-1,site-1:site+2])
            CA[t,site]= int(CARule(N)[neighbor])
    
    return CA      

```

```python
CA=CAEvolution(600,110)

plt.figure(figsize=(16,16))
plt.spy(CA)

plt.show()
```
